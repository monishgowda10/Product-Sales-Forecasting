# **Time Series Analysis and Forecasting**

# 2.1 Forecast at Enterprise Level (Overall Revenue)
We'll aggregate the data by month and forecast the overall revenue.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a 'TimeID' column for time series analysis
combined_data['TimeID'] = combined_data['InvoiceDate'].dt.to_period('M')

# Aggregate revenue by month
daily_revenue = combined_data.groupby('TimeID')['Revenue'].sum().reset_index()

# Convert 'TimeID' to datetime for plotting
daily_revenue['TimeID'] = daily_revenue['TimeID'].dt.to_timestamp()
```

```python
# Plot the time series
plt.figure(figsize=(12, 6))
sns.lineplot(x='TimeID', y='Revenue', data=daily_revenue)
plt.title('Monthly Revenue Over Time')
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.show()
```

```python
# Apply the Moving Average (3-month window)
window_size = 3  # 3-month moving average
daily_revenue['MovingAvg'] = daily_revenue['Revenue'].rolling(window=window_size).mean()

# Plot the original revenue and the moving average
plt.figure(figsize=(12, 6))

# Plot original revenue
sns.lineplot(x='TimeID', y='Revenue', data=daily_revenue, label='Original Revenue')

# Plot moving average
sns.lineplot(x='TimeID', y='MovingAvg', data=daily_revenue, label='3-Month Moving Average', color='red')

# Add title and labels
plt.title('Monthly Revenue Over Time with Moving Average')
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Fit the Holt-Winters Exponential Smoothing model
model = ExponentialSmoothing(daily_revenue['Revenue'], trend='add', seasonal='add', seasonal_periods=12)
fit = model.fit()

# Forecast for the next 12 months
forecast_periods = 12
forecast = fit.forecast(forecast_periods)

# Generate future dates for the forecast
future_dates = pd.date_range(start=daily_revenue['TimeID'].max(), periods=forecast_periods, freq='M')

# Plotting the results
plt.figure(figsize=(12, 6))

# Plot Actual Revenue
plt.plot(daily_revenue['TimeID'], daily_revenue['Revenue'], label='Actual Revenue', color='blue')

# Plot 3-Month Moving Average
plt.plot(daily_revenue['TimeID'], daily_revenue['MovingAvg'], label='3-Month Moving Average', color='orange', linestyle='--')

# Plot Forecasted Revenue
plt.plot(future_dates, forecast, label='Forecasted Revenue', color='green', linestyle='dashed')

# Title and labels
plt.title('Enterprise Revenue Forecast Model')
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

## **2.2 Forecast at Product Level**

We'll aggregate the data by product and month for forecasting. **bold text**

```python
# Aggregate revenue and quantity by product and month
product_monthly = combined_data.groupby(['StockCode', 'TimeID']).agg({'Revenue': 'sum', 'Quantity': 'sum'}).reset_index()

# Convert 'TimeID' to datetime for plotting
product_monthly['TimeID'] = product_monthly['TimeID'].dt.to_timestamp()

# Plot the time series for a specific product
product_code = '85123A'
product_data = product_monthly[product_monthly['StockCode'] == product_code]

plt.figure(figsize=(12, 6))
sns.lineplot(x='TimeID', y='Revenue', data=product_data)
plt.title(f'Monthly Revenue for Product {product_code} Over Time')
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.show()
```

```python
# Apply the Moving Average (e.g., 3-month window)
window_size = 3  # 3-month moving average
product_data['MovingAvg'] = product_data['Revenue'].rolling(window=window_size).mean()

# Plot the original revenue and the moving average
plt.figure(figsize=(12, 6))

# Plot original revenue
sns.lineplot(x='TimeID', y='Revenue', data=product_data, label='Original Revenue')

# Plot moving average
sns.lineplot(x='TimeID', y='MovingAvg', data=product_data, label=f'{window_size}-Month Moving Average', color='red')

# Add title and labels
plt.title(f'Monthly Revenue for Product 85123A Over Time with Moving Average')
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

# **Step 3: Building Forecasting Models**

We'll build different models for forecasting. For simplicity, we'll use moving average and neural networks (LSTM).

3.1 Moving Average Model

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Fit the model
model = ExponentialSmoothing(daily_revenue['Revenue'], trend='add', seasonal='add', seasonal_periods=12)
fit = model.fit()

# Forecast
forecast = fit.forecast(12)

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(daily_revenue['TimeID'], daily_revenue['Revenue'], label='Actual')
plt.plot(pd.date_range(start=daily_revenue['TimeID'].max(), periods=12, freq='M'), forecast, label='Forecast')
plt.title('Monthly Revenue Forecast')
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

3.2 Neural Network Model (LSTM)

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Prepare data for LSTM
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(daily_revenue[['Revenue']])

# Create sequences
def create_sequences(data, seq_length):
    sequences = []
    for i in range(len(data) - seq_length):
        sequences.append(data[i:i + seq_length])
    return np.array(sequences)

seq_length = 12
sequences = create_sequences(scaled_data, seq_length)
X, y = sequences[:, :-1], sequences[:, -1]

# Reshape for LSTM [samples, time steps, features]
X = X.reshape((X.shape[0], X.shape[1], 1))

# Build LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(seq_length - 1, 1)))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X, y, epochs=50, batch_size=32, verbose=1)

# Forecast
last_sequence = scaled_data[-seq_length:].reshape((1, seq_length, 1))
forecast = model.predict(last_sequence)

# Inverse transform the forecast
forecast = scaler.inverse_transform(forecast)
```

```python
#Step 4: Visualization
#We'll visualize the forecasts for both models.

# Forecast 12 periods with LSTM
forecast = []
last_sequence = scaled_data[-seq_length:].reshape((1, seq_length, 1))  # Start with the last sequence
for i in range(12):
    pred = model.predict(last_sequence)  # Predict one step ahead
    forecast.append(pred[0, 0])  # Add the prediction to the forecast list
    last_sequence = np.append(last_sequence[:, 1:, :], pred.reshape(1, 1, 1), axis=1)  # Update the last sequence

# Inverse transform the forecast
forecast = scaler.inverse_transform(np.array(forecast).reshape(-1, 1))

# Plot both forecasts together
plt.figure(figsize=(12, 6))
plt.plot(daily_revenue['TimeID'], daily_revenue['Revenue'], label='Actual')
plt.plot(pd.date_range(start=daily_revenue['TimeID'].max(), periods=12, freq='M'), forecast, label='LSTM Forecast') # Use forecasted values
plt.plot(pd.date_range(start=daily_revenue['TimeID'].max(), periods=12, freq='M'), fit.forecast(12), label='Moving Average Forecast')
plt.title('Monthly Revenue Forecast Comparison')
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```
